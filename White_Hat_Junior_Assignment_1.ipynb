```python
#as feaures are numerical as well as strings so require to convert them to numerical
#so we require some NLP related packages to preprocess the same
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
#here we are cleaning the text message by converting to small case, removing stop words like I, we, the etc., 
#doing some Lemmatization: convert words to their original form eg. talks--->talk
#also cleans the texts by removing ?, !, "",'' etc.
clean_text_utter = WordNetLemmatizer()
def clean_text(text_of_the_utterance):
    letters_only = re.sub("[^a-zA-Z]", " ", text_of_the_utterance)
    words = letters_only.lower().split()
    stops = set(stopwords.words("english")) 
    meaningful_words = [clean_text_utter.lemmatize(w) for w in words if not w in stops]   
    return( " ".join( meaningful_words ))
lines_characters_Data['clean_text'] = lines_characters_Data['text_of_the_utterance'].apply(clean_text)
#displaying some sample data after cleaning
lines_characters_Data[['text_of_the_utterance','clean_text']].sample(5)    
```

```python
#from this dataset now for training we are accomodating movie information on following columns as having multiple etries
#also combining them eg. character text, total no. of character Ids, character names etc.,
train = lines_characters_Data.groupby(['characterID', 'movieID', 'character_name', 'gender', 'position_in_credits']). \
            agg({'text_utter_length' : ['median'], 
                 'text_No_of_words' : ['median'],
                 'characterID' : ['count'],
                 'clean_text' : [lambda x : ' '.join(x)]
                })
train.columns = ["_".join(x) for x in train.columns.ravel()]#to write column name as per format
train.reset_index(inplace=True) #to resume the already available column indexes like characterID,movieID etc.
```

What is Normalization?
Normalization is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max scaling.

What is Standardization?
Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero and the resultant distribution has a unit standard deviation.

You can always start by fitting your model to raw, normalized and standardized data and compare the performance for best results.

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler
from sklearn.pipeline import FeatureUnion, Pipeline
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer, TfidfVectorizer
```

```python
#for training (fitting) and trasforming (normalization/statdardization) the data
class Converter(BaseEstimator, TransformerMixin):
    def fit(self, x, y=None):
        return self

    def transform(self, data_frame):
        return data_frame.values.ravel()
```

```python
#numerical features are collected in pipeline and applied scaling
numeric_features = ['position_in_credits', 'text_utter_length_median', 'text_No_of_words_median', 'characterID_count']

numeric_transformer = Pipeline(steps=[('scaler', MinMaxScaler())])
```

```python
#string feature is collected in pipeline and applied scaling
vectorizer_features = ['clean_text_<lambda>']
vectorizer_transformer = Pipeline(steps=[('con', Converter()),('tf', TfidfVectorizer())])
```

```python
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('vec', vectorizer_transformer, vectorizer_features)
    ])
```

```python
#importing LOGISTIC REGRESSION CLASSIFIER
from sklearn.linear_model import LogisticRegression
log_clf = Pipeline(steps=[('preprocessor', preprocessor),('classifier', LogisticRegression())])
```

```python
#actual training of classifier
log_clf.fit(X_train, y_train)
```

```python
#predicting on test input data
results=log_clf.predict(X_test)
```
